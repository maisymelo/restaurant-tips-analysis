## Análise de Gorjetas em Restaurantes

Neste estudo, analisei dados de um restaurante fictício com foco em:

- Comportamento de gorjetas por gênero e hábito de fumar
- Diferenças entre almoço e jantar
- Análise percentual entre valor da conta e gorjeta
- Visualizações estatísticas para entender padrões

**Principais insights:**
- Clientes não fumantes tendem a dar gorjetas maiores em média.
- O sábado é o dia com maior valor médio de contas.
- Existe correlação positiva entre o valor da conta e a gorjeta.

Este projeto faz parte da minha formação prática em Análise de Dados.



````python
# @title time vs size

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['time']
  ys = series['size']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df.sort_values('time', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('time')
_ = plt.ylabel('size')
````
